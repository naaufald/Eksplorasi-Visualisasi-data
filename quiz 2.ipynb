data berikut berisi list negara negara yang mengikuti pergeklaran olimpiade paris 2024 beserta perolehan medali yang mereka dapatkan

```python
import pandas as pd
quiz = pd.read_csv(r'D:\ALTO\olympics2024.csv')
```

```python
quiz
```

```python
selected_columns_1 = quiz.iloc[:, 0:2]
print(selected_columns_1)
```

```python
selected_columns_2 = quiz.iloc[:, 5:7]
print(selected_columns_2)
```

```python
data_iloc_1 = selected_columns_1.iloc[:,]
print(data_iloc_1)
```

```python
data_iloc_2 = selected_columns_2.iloc[:,]
data_iloc_2
```

```python
data_iloc_2.index = data_iloc_1.index
print(data_iloc_2)
```

```python
data_iloc_2.index = data_iloc_1.index
data_iloc_2['Rank'] = data_iloc_2.index + 1
data_iloc_2
```

Data di atas telah dilakukan penambahan index yang sama antara data_iloc_1 dan data_iloc_2 yaitu pada Rank, dimana, pada saat split pertama, data_iloc_2 tidak memiliki kolom rank.

```python
gabungan_data = pd.merge(data_iloc_1, data_iloc_2, left_index=False, right_index=False)
gabungan_data
```

data telah dirapikan

```python
gabungan_data.to_csv(r'D:\ALTO\merge_data.csv')
```
